```
import numpy as np
```

```
dataset = df.values
dataset
```

```
X = dataset[:,0:60].astype(float) #속성 중 정수와 실수가 섞여있다면 실수형태로 바꿔줘야 문제가 안생김
y = dataset[:,60] #X,Y구분
```

```
from sklearn.preprocessing import LabelEncoder
```

```
e = LabelEncoder()
e.fit(y)
y = e.transform(y)
```

```
from sklearn.model_selection import train_test_split # train 과 test를 쪼개줌 numpy,panda 도 있지만 리턴순서가 다 다름
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size=0.3) # 4개를 반환하기에 변수4개로 받아줌 7:3으로나눔
```

```
len(X_train), len(X_test)
```

```
len(y_train), len(y_test)
```

```
from keras.models import Sequential
from keras.layers.core import Dense
import tensorflow as tf
X.shape # (208.60) > input_dim = 60
```

```
y.shape # 208 = 1차원
```

```
model = Sequential()
model.add(Dense(24, input_dim=60, activation='relu'))
model.add(Dense(10, activation = 'relu'))
model.add(Dense(1, activation='sigmoid')) 
```

```
model.compile(loss='mean_squared_error',
              optimizer = 'adam',
              metrics=['accuracy'])
```

```
y
```

```
model.fit(X_train,y_train, epochs=200, batch_size=5)
```

```
model.evaluate(X,y) #학습데이터 성능 93%
```

```
model.evaluate(X_test,y_test) # 테스트데이터 성능 77%
```
