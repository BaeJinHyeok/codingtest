```python
import cv2
import numpy as np
from os import makedirs, listdir
from os.path import isdir, isfile, join
import sys
from PyQt5.QtWidgets import (QApplication, QWidget
, QLCDNumber, QDial, QPushButton, QVBoxLayout, QHBoxLayout, QLineEdit, QInputDialog)
from PyQt5.QtCore import QCoreApplication
from PyQt5.QtGui import QIcon
#import cv2
#import numpy as np
#from os import listdir
#from os.path import isdir, isfile, join

class MyApp(QWidget):

    def __init__(self):
        super().__init__()
        self.initUI()

    def initUI(self):
        btn1 = QPushButton('Capture', self)
        
        btn1.clicked.connect(self.showDialog)

        self.le = QLineEdit(self)
        self.le.move(120, 35)



        btn2 = QPushButton('Model Training', self)
        btn3 = QPushButton('Recognize', self)
        btn = QPushButton('Quit', self)
        
        
        hbox = QHBoxLayout()
        hbox.addWidget(btn1)
        hbox.addWidget(btn2)
        hbox.addWidget(btn3)
        hbox.addWidget(btn)
        
        vbox = QVBoxLayout()
        vbox.addLayout(hbox)
        self.setLayout(vbox)




        #btn = QPushButton('Quit', self)
        #btn.move(50, 50)
        btn.resize(btn.sizeHint())
        btn1.resize(btn1.sizeHint())
        btn2.resize(btn2.sizeHint())
        btn3.resize(btn3.sizeHint())
        
        btn.clicked.connect(self.close)
            
        btn1.clicked.connect(self.capTure)
        btn2.clicked.connect(self.traiNing)
        btn3.clicked.connect(self.recoGnize)
        
        self.setWindowTitle('Face')
        self.setWindowIcon(QIcon('face'))
        self.setGeometry(300, 300, 300, 200)
        self.show()
        
    def showDialog(self):
        global face_id
        global face_name
        face_id, ok = QInputDialog.getText(self, 'Student Number', '학번을 입력하세요:')
        face_name, ok = QInputDialog.getText(self, 'Student Name', '이름을 입력하세요:')
        if ok:
            self.le.setText(str(face_id))
            self.le.setText(str(face_name))
            
        
        
#캡쳐버튼


    def capTure(self):
        face_dirs = 'faces/'
        face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')
        
        def showDialog(self):
            print("학번 :" " \n 이름 :")
            
          

       
        def face_extractor(img):
            gray = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
            faces = face_classifier.detectMultiScale(gray,1.3,5)
            if faces is():
                return None
            for(x,y,w,h) in faces:
                cropped_face = img[y:y+h, x:x+w]
                
            return cropped_face

        def take_pictures(face_id):
            if not isdir(face_dirs+face_id):
                makedirs(face_dirs+face_id)

                   
            cap = cv2.VideoCapture(0)
            count = 0
           

            while True:
                ret, frame = cap.read()
                if face_extractor(frame) is not None:

                    count+=1
                    face = cv2.resize(face_extractor(frame),(200,200))
                    face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

                    file_name_path = face_dirs + face_id + '/' +face_name+str(count)+'.jpg'
                    cv2.imwrite(file_name_path,face)

                    cv2.putText(face,str(count),(50,50),cv2.FONT_HERSHEY_COMPLEX,1,(0,255,0),2)
                    cv2.imshow('Face recognition attendance',face)
                else:
                    print("Face not Found")
                    pass

                if cv2.waitKey(1)==13 or count==100:
                    break

            cap.release()
            cv2.destroyAllWindows()
            print(' 캡쳐 완료')


        if __name__ == "__main__":
            take_pictures(face_id)
           


 #학습버튼

    def traiNing(self):
        
        #face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')    

        def train(name):
            data_path = 'faces/' + name + '/'
            face_pics = [f for f in listdir(data_path) if isfile(join(data_path,f))]

            Training_Data, Labels = [], []

            for i, files in enumerate(face_pics):
                image_path = data_path + face_pics[i]
                images = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)
                if images is None:
                    continue    
                Training_Data.append(np.asarray(images, dtype=np.uint8))
                Labels.append(i)
            if len(Labels) == 0:
                print("There is no data to train.")
                return None
            Labels = np.asarray(Labels, dtype=np.int32)
            model = cv2.face.LBPHFaceRecognizer_create()
            model.train(np.asarray(Training_Data), np.asarray(Labels))
            print(name + " : Model Training Complete")

            return model

        def trains():
            data_path = 'faces/'
            model_dirs = [f for f in listdir(data_path) if isdir(join(data_path,f))]
            global models
            models = {}
            for model in model_dirs:
                print('model :' + model)
                result = train(model)
                if result is None:
                    continue
                print('model2 :' + model)
                models[model] = result

            return models    

 #       def face_detector(img, size = 0.5):
  #              gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
   #             faces = face_classifier.detectMultiScale(gray,1.3,5)
    #            if faces is():
     #               return img,[]
      #          for(x,y,w,h) in faces:
       #             cv2.rectangle(img, (x,y),(x+w,y+h),(0,255,0
        #                                               ),2)
         #           roi = img[y:y+h, x:x+w]
          #          roi = cv2.resize(roi, (200,200))
           #     
            #    return img,roi   
        if __name__ == "__main__":
            
            
            models = trains()
            
            #run(models)
           
    
#인식버튼
    def recoGnize(self):
        face_classifier = cv2.CascadeClassifier('haarcascade_frontalface_default.xml')    

        def face_detector(img, size = 0.5):
                gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
                faces = face_classifier.detectMultiScale(gray,1.3,5)
                if faces is():
                    return img,[]
                for(x,y,w,h) in faces:
                    cv2.rectangle(img, (x,y),(x+w,y+h),(0,255,0
                                                       ),2)
                    roi = img[y:y+h, x:x+w]
                    roi = cv2.resize(roi, (200,200))
                
                return img,roi   
            
        def run(models):    
            cap = cv2.VideoCapture(0)

            while True:
                ret, frame = cap.read()
                image, face = face_detector(frame)
                try:            
                    min_score = 999       
                    min_score_name = ""   


                    face = cv2.cvtColor(face, cv2.COLOR_BGR2GRAY)

                    for key, model in models.items():
                        result = model.predict(face)                
                        if min_score > result[1]:
                            min_score = result[1]
                            min_score_name = key
                           

                    if min_score < 500: 
                        confidence = int(100*(1-(min_score)/300))
                        display_string = str(confidence)+'% Confidence ' + min_score_name
                    cv2.putText(image,display_string,(100,120), cv2.FONT_HERSHEY_COMPLEX,1,(0, 255, 0),2)
                    if confidence > 75:
                        cv2.putText(image, "Student : " + min_score_name, (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (0, 255, 0), 2)
                        cv2.imshow('Face recognition attendance', image)
                    else:
                        cv2.putText(image, "Unknown", (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 0, 0), 2)
                        cv2.imshow('Face recognition attendance', image)
                except:
                    cv2.putText(image, "Face Not Found", (250, 450), cv2.FONT_HERSHEY_COMPLEX, 1, (255, 255, 255), 2)
                    cv2.imshow('Face recognition attendance', image)
                    pass
                if cv2.waitKey(1)==13:
                    break

        if __name__ == "__main__":
            run(models)

if __name__ == '__main__':
    app = QApplication(sys.argv)
    ex = MyApp()
    sys.exit(app.exec_())
```
