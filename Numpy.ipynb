```python
#배열 속성 정보
def array_info(array):
    print(array)
    print("ndim:", array.ndim) # 1차원
    print("shape:", array.shape) #5개의 엘리먼트
    print("dtype:", array.dtype) #인트
    print("size:", array.size) # 5
    print("itemsize:", array.itemsize) # 각 하나의 아이템이 가지는 사이트 8nyte =>int32 이므로 4
    print("nbytes:", array.nbytes) #전체 바이츠 4*5 = 20
    print("strides:", array.strides) #다음 엘리먼트(차원)으로 가기위한 사이즈 = 4 가 필요
```
